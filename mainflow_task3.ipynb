```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('householdtask3.csv')
```

```python
print(df.head())
```

```python
plt.figure(figsize=(12, 6))
plt.bar(df['year'], df['tot_hhs'], color='blue', alpha=0.6, label='Total Households')
plt.xlabel('Year')
plt.ylabel('Total Households')
plt.title('Total Households by Year')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['year'], df['income'], marker='o', color='red', linestyle='-', label='Average Income')
plt.plot(df['year'], df['expenditure'], marker='o', color='green', linestyle='-', label='Average Expenditure')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Average Income and Expenditure by Year')
plt.legend()
plt.show()
```
